```python
import mplfinance as mpf

data = yf.download('0050.TW', start='2025-08-01' , auto_adjust=False)
data.columns = data.columns.get_level_values(0)  # 移除多層欄位名稱
data.tail(5)
```

```python
# mpf.plot(data) # 基本的K線圖
# mpf.plot(data, type='candle') # 蠟燭圖

mpf.available_styles()  # 查看可用的樣式名稱
# mpf.plot(data, type='candle', style='yahoo') # 使用yahoo風格
mpf.plot(data, type='candle', style='binance') # 使用binance風格
```

```python
data['5MA'] = data.rolling(5)['Close'].mean() # 計算5日均線
data['20MA'] = data.rolling(20)['Close'].mean() # 計算20日均線
addps = [] # 建立一個空的列表來存放均線
addps.append(mpf.make_addplot(data['5MA'])) # 把5日均線加入列表
addps.append(mpf.make_addplot(data['20MA'])) # 把20日均線加入列表
mpf.plot(data, type='candle', style='yahoo', addplot=addps) # 把均線加到K線圖上
```

```python
data.head(25)
```

```python
data['floor_now'] = data.rolling(3)['Low'].min() 
data['floor'] = data['floor_now'].shift()
data['ceil_now'] = data.rolling(3)['High'].max()
data['ceil'] = data['ceil_now'].shift()
addps = []
addps.append(mpf.make_addplot(data['floor']))
addps.append(mpf.make_addplot(data['ceil']))
mpf.plot(data, type='candle', style='yahoo', addplot=addps)
```
